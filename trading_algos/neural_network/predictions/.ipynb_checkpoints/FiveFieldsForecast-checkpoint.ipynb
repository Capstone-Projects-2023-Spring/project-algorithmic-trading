```python
import os
import yaml
import pandas as pd
import numpy as np
from pathlib import Path
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()

os.chdir("../scripts")

import preprocess, train, inference, interpret
from config import config_dict

import pandas_market_calendars as mcal

from datetime import datetime,timedelta
from dateutil.relativedelta import relativedelta

from alpha_vantage.timeseries import TimeSeries 
from pathlib import Path
```

```python
with open("../model/params.yaml", "r") as params_file:
    params = yaml.safe_load(params_file)

data_dir = params['data_dir']
```

```python
ts = TimeSeries(key=config_dict["av"]["a_k"], output_format=config_dict["av"]["format"])
data, meta_data = ts.get_daily_adjusted(config_dict["av"]["symbol"], outputsize=config_dict["av"]["outputsize"])

data = data.iloc[::-1]
data = data[config_dict["av"]["fields"]]
mapping = {data.columns[0]:'open', data.columns[1]: 'close', data.columns[2]:'low', data.columns[3]: 'high',data.columns[4]:'volume'}
data = data.rename(columns=mapping)

os.chdir("..")
filepath = Path("data/SPY.csv")
data.to_csv(filepath)
```

```python
os.chdir('scripts')
file_name = "SPY.csv"
data = preprocess.load_data(file_name)
data.columns = data.columns.str.capitalize()
data.tail(3)
```

```python
n = len(data.index)
date_data = data['Date'].to_numpy()
next_weeks = mcal.date_range(mcal.get_calendar('NYSE').schedule(start_date=date_data[n-1], end_date=(datetime.strptime(date_data[n-1],'%Y-%m-%d')+relativedelta(months=1)).strftime('%Y-%m-%d')), frequency='1D')
next_weeks = [date.strftime('%Y-%m-%d') for date in next_weeks]
next_weeks = next_weeks[1:]
empty = np.ones(len(next_weeks))
```

```python
temp_dict = {
    'Date': next_weeks,
    'Open': empty,
    'Close': empty,
    'Low': empty,
    'High': empty,
    'Volume': empty,
}
temp_df = pd.DataFrame(temp_dict)
data = pd.concat([data,temp_df],ignore_index = True)
data.reset_index()
data
```

```python
print(n)
print(len(data.index))
train_df, test_df = preprocess.prep_data(df=data, split_index=n-1,validate = False)
```

```python
train_df.shape, test_df.shape
```

```python
sequence_length = len(test_df)
train_batch_size = config_dict["training"]["train_batch_size"]
test_batch_size = config_dict["training"]["test_batch_size"]
n_epochs = config_dict["training"]["num_epoch"]
n_epochs_stop = 10
label_name = 'Close'
```

```python
test_df
```

```python
hist = train.train_model(train_df, test_df, label_name, sequence_length, train_batch_size, test_batch_size, n_epochs, n_epochs_stop)
```

```python
predictions_descaled, labels_descaled = inference.predict(df=test_df,
                                                          label_name='Close',
                                                          sequence_length=sequence_length)

predictions_descaled.shape, labels_descaled.shape
```

```python
# plot predictions vs true values
df_pred = pd.DataFrame()
df_pred['predicted'] = predictions_descaled
df_pred['residual'] = labels_descaled - predictions_descaled
df_pred.plot(figsize=(12,6), cmap='viridis')
plt.xticks(rotation=45)
#plt.xlim(0,300)
plt.xlabel('days')
plt.ylabel('price')
plt.legend()
plt.show()
```
